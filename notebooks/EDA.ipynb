# Data Cleaning and Preprocessing

In this step, we will:
- **Handle Missing Values:**
  Impute missing values using appropriate strategies for numerical (median) and categorical (mode) features. We will also consider removing columns if they contain too many missing values.

- **Detect and Handle Outliers:**
  Visualize distributions with histograms and boxplots to identify anomalies and decide whether to transform or remove extreme outliers.

- **Data Type Conversion:**
  Convert columns to appropriate types (e.g., convert certain text columns to categorical).

- **Documentation:**
  All cleaning steps, assumptions, and transformations will be clearly recorded.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from train.csv
train_data_path = '../data/raw/house_prices_data/train.csv'
df_train = pd.read_csv(train_data_path)
```

## Handling Missing Values

We will impute missing values for numerical columns with the median and for categorical columns with the mode. Additionally, if a column has more than 50% missing values, we may decide to drop it.


```python
# Identify numerical and categorical columns
num_cols = df_train.select_dtypes(include=[np.number]).columns
cat_cols = df_train.select_dtypes(include=['object']).columns

print(f"num_cols : {num_cols}")
print(f"cat_cols : {cat_cols}")
```

```python
# Impute missing values for numerical columns using the median
df_train[num_cols] = df_train[num_cols].fillna(df_train[num_cols].median())

# Impute missing values for categorical columns using the mode
df_train[cat_cols] = df_train[cat_cols].fillna(df_train[cat_cols].mode().iloc[0])
```

```python
# Drop columns with more than 50% missing values (if any)
threshold = 0.5
cols_to_drop = df_train.columns[df_train.isnull().mean() > threshold]

if len(cols_to_drop) > 0:
    print("Dropping columns with more than 50% missing values:", list(cols_to_drop))
    df_train = df_train.drop(columns=cols_to_drop)
```

## Outlier Detection and Handling

We visualize the distributions of numerical features using histograms and boxplots to identify any anomalies. For example, we will inspect the distribution of the 'SalePrice' column.
If extreme outliers are detected, we can remove or transform them. Here, we'll demonstrate the Interquartile Range (IQR) method to remove outliers in 'SalePrice'.


```python
# Plot histograms for numerical columns
for col in ['SalePrice', 'GrLivArea']:  # Adjust columns as needed
    plt.figure(figsize=(8, 4))
    sns.histplot(df_train[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Boxplot for 'SalePrice' to visualize outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_train['SalePrice'])
plt.title('Boxplot of SalePrice')
plt.xlabel('SalePrice')
plt.show()
```

```python
# Remove outliers in 'SalePrice' using the IQR method
Q1 = df_train['SalePrice'].quantile(0.25)
Q3 = df_train['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
initial_count = df_train.shape[0]
df_train = df_train[(df_train['SalePrice'] >= lower_bound) & (df_train['SalePrice'] <= upper_bound)]
final_count = df_train.shape[0]
print(f"Removed {initial_count - final_count} outliers from SalePrice.")
```

## Data Type Conversion

Convert columns to appropriate data types. For example, if any columns represent categorical data as strings, we can convert them to the 'category' type.


```python
# Convert a sample column to categorical, if applicable
if 'MSZoning' in df_train.columns:
    df_train['MSZoning'] = df_train['MSZoning'].astype('category')

# Additional conversions can be added here as needed
print("Data types after conversion:")
print(df_train.dtypes)
```

- **Boxplots:** to detect the spread and potential outliers.

## Bivariate and Multivariate Analysis

Next, we explore relationships between features:
- **Correlation Heatmap:** to identify linear relationships among numerical variables.
- **Pair Plots:** to examine pairwise relationships.
- **Scatter Plot:** to specifically explore the relationship between `GrLivArea` and `SalePrice`.
- **Bar Plot:** to visualize the average `SalePrice` grouped by a categorical variable (e.g., `OverallQual`).

# Feature Engineering

In this step, we will enhance our dataset by:
1. **Creating New Features:**
   - Deriving new features based on domain insights.
   - For example, creating a `TotalSF` variable as the sum of `TotalBsmtSF`, `1stFlrSF`, and `2ndFlrSF` (if available).
   - Creating interaction features like `OverallGrade` by multiplying `OverallQual` and `OverallCond`.

2. **Encoding:**
   - Transforming categorical features into numerical form using one-hot encoding.

3. **Scaling:**
   - Applying standardization to numerical features using a scaler.

4. **Feature Importance:**
   - Fitting a simple linear regression model to identify features that have the most impact on the target variable (e.g., `SalePrice`).

All these steps will help in preparing the data for any downstream modeling tasks.


```python
import pandas as pd
import numpy as np

# Load the cleaned dataset (assuming it's already cleaned from previous steps)
df = pd.read_csv('..//data/raw/house_prices_data/train.csv')

# Create new feature 'TotalSF' if the required columns exist
required_cols = ['TotalBsmtSF', '1stFlrSF', '2ndFlrSF']
if set(required_cols).issubset(df.columns):
    df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']

# Create an interaction feature 'OverallGrade' if 'OverallQual' and 'OverallCond' exist
if {'OverallQual', 'OverallCond'}.issubset(df.columns):
    df['OverallGrade'] = df['OverallQual'] * df['OverallCond']

# Display the new features
print("Newly Created Features:")
print(df[['TotalSF', 'OverallGrade']].head())
```

## Encoding Categorical Variables

We need to transform categorical variables into a numerical format. Here, we use one-hot encoding.


```python
# Identify categorical columns (of type 'object' or 'category')
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Apply one-hot encoding using pd.get_dummies, dropping the first level to avoid multicollinearity
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

print("Data after One-Hot Encoding:")
print(df_encoded.head())
```

## Scaling Numerical Features

Standardizing numerical features can be beneficial for many modeling techniques.
We will use Scikit-learn's StandardScaler to standardize the numerical columns, excluding the target variable `SalePrice`.


```python
from sklearn.preprocessing import StandardScaler

# Select numeric features from the encoded dataframe
numeric_features = df_encoded.select_dtypes(include=[np.number]).columns.tolist()

# Remove the target variable 'SalePrice' from scaling (if present)
if 'SalePrice' in numeric_features:
    numeric_features.remove('SalePrice')

# Initialize the scaler
scaler = StandardScaler()

# Scale the numeric features
df_encoded_scaled = df_encoded.copy()
df_encoded_scaled[numeric_features] = scaler.fit_transform(df_encoded_scaled[numeric_features])

print("Data after Scaling (first 5 rows):")
print(df_encoded_scaled.head())
```

## Feature Importance Evaluation

A simple linear regression model can give us an idea of which features are most impactful.
We fit a linear regression model with `SalePrice` as the target, then inspect the coefficients (using absolute values for ranking).


```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Prepare the data: use all features except 'SalePrice' as predictors (X) and SalePrice as the target (y)
X = df_encoded_scaled.drop(columns=['SalePrice'])
y = df_encoded_scaled['SalePrice']

# Check for any missing values in predictors and target
print("Missing values in X:", X.isnull().sum().sum())
print("Missing values in y:", y.isnull().sum().sum())

# Option 1: Fill missing values in X with 0
X = X.fillna(0)

# Option 2: Drop rows with missing values
# X = X.dropna()
# y = y[X.index]

# Fit a simple linear regression model
model = LinearRegression()
model.fit(X, y)

# Extract coefficients and create a DataFrame for feature importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})
feature_importance['AbsoluteCoefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='AbsoluteCoefficient', ascending=False)

print("Top 10 Feature Importances:")
print(feature_importance.head(10))

# Plot the top 10 features by absolute coefficient value
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance.head(10), x='AbsoluteCoefficient', y='Feature')
plt.title('Top 10 Feature Importances (Absolute Coefficient Values)')
plt.xlabel('Absolute Coefficient')
plt.ylabel('Feature')
plt.show()
```
